```python
def getPOI(s) :
    return s.split("/")[0]

def getStreet(s) :
    return s.split("/")[1]

def split(s) :
    return s.split(" ")
```

```python
df["POI"] = df["POI/street"].apply(getPOI)
df["Street"] = df["POI/street"].apply(getStreet) 
```

```python
df
```

```python
# Empty POI
len(df[df['POI'] == '']) 
```

```python
# Empty Street
len(df[df['Street'] == '']) 
```

```python
df["SplitStreet"] = df["Street"].apply(split)
df["SplitPOI"] = df["POI"].apply(split)
df["SplitRaw"] = df["raw_address"].apply(split)
```

```python
def f(words, sentence): 
    res = []
    for word in words :
        if sentence.find(word) == -1 :
            res.append(word)
    if len(res) == 0 :
        return False
    return res
```

```python
df['TransformedPOI'] = df.apply(lambda x: f(x.SplitPOI, x.raw_address), axis=1)
df['TransformedStreet'] = df.apply(lambda x: f(x.SplitStreet, x.raw_address), axis=1)
```

```python
df[df["TransformedPOI"] != False]
```

```python
transform_poi = {}
for transformed, address  in zip(df["TransformedPOI"], df["SplitRaw"]) :
    if transformed :
        for t in transformed :
            for a in address :
                if a == t[:len(a)] :
                    transform_poi[a] = t
                    break

len(transform_poi)
```

```python
df[df["TransformedStreet"] != False]
```

```python
transform_street = {}
for transformed, address  in zip(df["TransformedStreet"], df["SplitRaw"]) :
    if transformed :
        for t in transformed :
            for a in address :
                if a == t[:len(a)] :
                    transform_street[a] = t
                    break

len(transform_street)
```

```python
def preprocess(words) :
    res = []
    for word in words :
        if word in transform_poi :
            res.append(transform_poi[word])
        elif word in transform_street :
            res.append(transform_street[word])
        else :
            res.append(word)

    return " ".join(res) 
    
df["Raw"] = df["SplitRaw"].apply(preprocess)
```
