```python
def getPOI(s) :
    return s.split("/")[0]

def getStreet(s) :
    return s.split("/")[1]

def split(s) :
    return s.split(" ")
```

```python
df["POI"] = df["POI/street"].apply(getPOI)
df["Street"] = df["POI/street"].apply(getStreet) 
```

```python
df["SplitStreet"] = df["Street"].apply(split)
df["SplitPOI"] = df["POI"].apply(split)
df["SplitRaw"] = df["raw_address"].apply(split)
```

```python
def f(words, sentence): 
    res = []
    for word in words :
        if sentence.find(word) == -1 :
            res.append(word)
    if len(res) == 0 :
        return False
    return res
```

```python
df['TransformedPOI'] = df.apply(lambda x: f(x.SplitPOI, x.raw_address), axis=1)
df['TransformedStreet'] = df.apply(lambda x: f(x.SplitStreet, x.raw_address), axis=1)
```

```python
ndf = df[df["TransformedPOI"] == False]
ndf = ndf[ndf["TransformedStreet"] == False]
ndf
```

```python
# ndf = ndf.replace("", np.NaN)
cleandf = ndf.dropna()
cleandf = cleandf[["raw_address","POI", "Street"]]
cleandf
```

```python
def comma_preprocess(s) :
    s = s.replace("\t", " ")
    s = re.sub(r'([,/\\]+)', r' \1 ', s) # 
    s = re.sub(' +', ' ', s) # change multiple spaces into one space
    return s.strip()

s = "raya samb gede, 299 toko bb kids"
s = "jl. amd, komplek borneo lestari, blok 2, no. 30"
comma_preprocess(s)
```

```python
cleandf["Address"] = cleandf["raw_address"].apply(comma_preprocess)
```

```python
cleandf
```

```python
def create_label(s, poi, street) :
    tokens = s.split(" ")
    n = len(tokens)
    tp = poi.split(" ")
    ts = street.split(" ")
    tags = []
    i = 0
    while i < n :
        if len(tp) > 0 and tokens[i] == tp[0] :
            j = 1
            while j < len(tp) :
                if i+j < n and tokens[i+j] == tp[j]:
                    j += 1
                else :
                    break
            if j == len(tp) :
                tags.append("B-ORGANISATION")
                for k in range(1, j) :
                    tags.append("I-ORGANISATION")
                i += j
            else :
                tags.append("O")
                i += 1
        elif len(ts) > 0 and tokens[i] == ts[0] :
            j = 1
            while j < len(ts) :
                if i+j < n and tokens[i+j] == ts[j] :
                    j += 1
                else :
                    break
            if j == len(ts) :
                tags.append("B-PLACE")
                for k in range(1, j) :
                    tags.append("I-PLACE")
                i += j
            else :
                tags.append("O")
                i += 1                
        else :
            tags.append("O")
            i += 1
        
#         print(tokens)
#         print(tags)

    if tokens[-1] != "." :
        tokens.append(".")
        tags.append("O")
    
    res = ""
#     print(tokens)
#     print(tags)
    for token, tag in zip(tokens, tags) :
        res += "{}\t{}\n".format(token, tag)
    return res
```

```python
s = "raya samb gede , 299 toko bb kids"
poi = "toko bb kids"
street = "raya samb gede"
res = create_label(s, poi, street)
print(res)
```

```python
cleandf["Label"] = cleandf.apply(lambda x: create_label(x.Address, x.POI, x.Street), axis=1)
```

```python
cleandf
```

```python
final = cleandf.sample(frac=1, random_state=0).reset_index(drop=True)
final
```

```python
n = len(final)
nt = int(0.95 * n)
train = final[:nt+1]
test = final[nt:]
```

```python
def save_file(filename, arr) :
    f = open(filename, "w+",  encoding='utf-8')
    for s in arr :
        f.write(s.encode("ascii", "ignore").decode())
        f.write("\n")
    f.close()
```

```python
train_path = data_path + "bert-fine-tune/train.txt"
validation_path = data_path + "bert-fine-tune/validation.txt"

train_data = train["Label"].values.tolist()
test_data = test["Label"].values.tolist()


ntry = None
# ntry = 1000
if ntry :
    train_data = train_data[:ntry]
    test_data = test_data[:ntry]

save_file(train_path, train_data)
save_file(validation_path, test_data)
```
