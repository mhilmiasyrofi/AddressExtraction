```python
def getPOI(s) :
    return s.split("/")[0]

def getStreet(s) :
    return s.split("/")[1]

def split(s) :
    return s.split(" ")
```

```python
df["POI"] = df["POI/street"].apply(getPOI)
df["Street"] = df["POI/street"].apply(getStreet) 
```

```python
# Empty Street
len(df[df['Street'] == '']) 
```

```python
df["SplitStreet"] = df["Street"].apply(split)
df["SplitPOI"] = df["POI"].apply(split)
df["SplitRaw"] = df["raw_address"].apply(split)
```

```python
def f(words, sentence): 
    res = []
    for word in words :
        if sentence.find(word) == -1 :
            res.append(word)
    if len(res) == 0 :
        return False
    return res
```

```python
df['TransformedPOI'] = df.apply(lambda x: f(x.SplitPOI, x.raw_address), axis=1)
df['TransformedStreet'] = df.apply(lambda x: f(x.SplitStreet, x.raw_address), axis=1)
```

```python
df[df["TransformedPOI"] != False]
```

```python
transform_poi = {}
for transformed, address  in zip(df["TransformedPOI"], df["SplitRaw"]) :
    if transformed :
        for t in transformed :
            for a in address :
                if a == t[:len(a)] :
                    transform_poi[a] = t
                    break

len(transform_poi)
```

```python
df[df["TransformedStreet"] != False]
```

```python
transform_street = {}
for transformed, address  in zip(df["TransformedStreet"], df["SplitRaw"]) :
    if transformed :
        for t in transformed :
            for a in address :
                if a == t[:len(a)] :
                    transform_street[a] = t
                    break

len(transform_street)
```

```python
transform_poi["jl."]
```

```python
transform_street["jl."]
```

```python
def preprocess(words) :
    res = []
    for word in words :
        if word in transform_poi :
            res.append(transform_poi[word])
        elif word in transform_street :
            res.append(transform_street[word])
        else :
            res.append(word)

    return " ".join(res) 
    
df["Raw"] = df["SplitRaw"].apply(preprocess)
```

## Street Feature Engineering

```python
def isContain(s, cek) :
    return cek in s
```

```python
# street_tag = ["jl", "jalan", "gang", "gg.", "raya"]
street_tag = ["raya", "gg.", "jl.", "jl"]
second_street_tag = ["gang", "jalan", "jend", "indah", "timur", "barat", "haji", "utara", "selatan"]
street_number = ["i", "ii", "iii", "iv", "v", "vi", "vii", "viii", "ix"]
# for tag in street_tag + street_number :
#     df[tag] = df["SplitStreet"].apply(isContain, args=(tag,))
#     print(tag, ": ", len(df[df[tag] == True]))
```

```python
df
```

```python
# df[df["ii"] == True]
```

```python
def extract_street(s) :
    
    for t in street_tag :
        loc = s.find(t)
        if loc != -1 :
            left = 0
            i = 0
            while i < loc :
                if s[i] == "," :
                    left = i + 1
                i += 1
            right = len(s) - 1
            while i < len(s) :
                if s[i] == "," :
                    right = i - 1
                    break
                if s[i].isnumeric() :
                    right = i - 1
                    if s[i-3:i-1] == "no" :
                        right -= 4
                    break
                i += 1
            res = s[left:right+1]
            return res.strip()
    
    
    for t in street_number :
        middle_t = " " + t + " "
        loc = s.find(middle_t)
        if loc == -1 :
            loc = s.find(" " + t + ",")
        if loc != -1 :
            left = 0
            i = 0
            while i < loc :
                if s[i] == "," :
                    left = i + 1
                i += 1
            right = loc + len(t)
            res = s[left:right+1]
            return res.strip()
        
    for t in second_street_tag :
        loc = s.find(t)
        if loc != -1 :
            left = 0
            i = 0
            while i < loc :
                if s[i] == "," :
                    left = i + 1
                i += 1
            right = len(s) - 1
            while i < len(s) :
                if s[i] == "," :
                    right = i - 1
                    break
                if s[i].isnumeric() :
                    right = i - 1
                    if s[i-3:i-1] == "no" :
                        right -= 4
                    break
                i += 1
            res = s[left:right+1]
            return res.strip()

    return ""
```

## POI Feature Engineering

```python
poi_tag = ["toko", "warung", "pt", "jaya", "masjid", "taman", "negeri", "perumahan", "sd", "motor", "komplek", "cell","kantor","pasar", "indonesia", "gudang", "rumah", "asrama", "stasiun"]
# for tag in poi_tag :
#     df[tag] = df["SplitPOI"].apply(isContain, args=(tag,))
#     print(tag, ": ", len(df[df[tag] == True]))
```

```python
# df[df["toko"] == True]
```

```python
def extract_poi(s) :

    for t in poi_tag :
        loc = s.find(t)
        if loc != -1 :
            left = 0
            i = 0
            while i < loc :
                if s[i] == "," or s[i] == ".":
                    left = i + 1
                i += 1
            right = len(s) - 1
            while i < len(s) :
                if s[i] == "," or s[i] == ".":
                    right = i - 1
                    break
                if s[i].isnumeric() :
                    right = i - 1
                    break
                i += 1
            res = s[left:right+1]
            return res.strip()
    


    return ""
```
